# Exploring an Existing `rubicon-ml` Project

[This GitHub repo](https://github.com/ryanSoley/experimenting-with-rubicon) contains
a simple classification model in `my_model/my_model.py`. Over three commits, three
different types of models were used to attempt to classify species of penguins by
some physical characteristics. The `rubicon-ml` logs generated and written while
running the code in `my_model.py` at each of the three commits have been copied
locally into this Binder session. In a more realistic use case, you may be reading
shared `rubicon-ml` logs from S3.

First, you'll have to infer this Binder session's proxy information (This first step
is only required when running the dashboard from a Binder session).

```python
from jupyter_dash import JupyterDash


JupyterDash.infer_jupyter_proxy_config()
```

And now you can launch the dashboard! Click the link in the output below
to launch the dashboard in a new browser window.
By using the dashboard to explore the `rubicon-ml` logs generated by the
`my_model.py` runs, there are a number of questions you'll be able to answer:
* What were the input parameters and output metrics of each model run?
* What type of model did each run use?
* Where is the source code that was used to generate each model run's logs?
* Which model and input parameters produced the best results?

```python
import os

from rubicon.ui import Dashboard


Dashboard(persistence="filesystem", root_dir=f"{os.getcwd()}/rubicon-root").run_server(debug=True)
```

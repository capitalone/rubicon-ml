# Exploring an Existing `rubicon_ml` Project

[This GitHub repo](https://github.com/ryanSoley/experimenting-with-rubicon) contains
a simple classification model in `my_model/my_model.py`. Over three commits, three
different types of models were used to attempt to classify species of penguins by
some physical characteristics. The `rubicon_ml` logs generated and written while
running the code in `my_model.py` at each of the three commits have been copied
locally into this Binder session. In a more realistic use case, you may be reading
shared `rubicon_ml` logs from S3.

If you're running this from a Binder session, you'll have to first infer its proxy
information (Otherwise you can skip running this cell).

```python
from jupyter_dash import JupyterDash


JupyterDash.infer_jupyter_proxy_config()
```

Then launch the dashboard by running the cell below! Click the link in the output
to launch the dashboard in a new browser window and select the **Classifying
Penguins** project from the sidebar. By using the dashboard to explore the
`rubicon_ml` logs generated by the `my_model.py` runs, there are a number of
questions you'll be able to answer:
* What were the input parameters and output metrics of each model run?
* What type of model did each run use?
* Where is the source code that was used to generate each model run's logs?
* Which model and input parameters produced the best results?

```python
import os

from rubicon_ml.ui import Dashboard


Dashboard(
    persistence="filesystem",
    root_dir=f"{os.getcwd()}/rubicon-root",
).run_server()
```
